```python
final= pd.DataFrame()
for j in range(1,16):
    url = requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j))
    url.text
    soup=bs4.BeautifulSoup(url.text, 'html')
    
    company_content= soup.find_all('div', class_='company-content-wrapper')
    
    name=[]
    ratings=[]
    reviews=[]
    ctype=[]
    hq=[]
    old=[]
    employees=[]
    about=[]

    for i in company_content :
            name.append(i.find('h2').text.strip())
            ratings.append(i.find('p', class_='rating').text.strip())
            reviews.append(i.find('a', class_='review-count').text.strip())
            ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
            try:
                 hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())
            except:
                hq.append(np.nan)
            try:
                 old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
            except:
                old.append(np.nan)
            try:
                employees.append(i.find_all('p', class_='infoEntity')[3].text.strip())
            except:
                employees.append(np.nan)
            about.append(i.find('p', class_='description body-small').text.strip())
          
    
    d={'Companyname':name, 'Ratings':ratings, 'Reviews':reviews, 'CompanyType': ctype, 'Headquaters':hq, 'Old': old, 'Employees':employees , 'About':about}
    df= pd.DataFrame(d)
    
    final= final.append(df)
```
